### 7. ¿Cuántas marcas blancas tiene mi competencia? , ¿estamos mejor o peor posicionados según precio y variedad?

* Si queremos saber cuántos productos bajo su marca blanca tiene nuestra competencia, primero debemos separar nuestro DataFrame por Supermercado. Lo hacemos a través de la columna "name", para así obtener los productos que corresponden a cada Marca Blanca.
* Luego debemos contabilizar los productos de cada marca blanca, para ello debemos filtrar los valores únicos para evitar contar artículos duplicados.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.float_format', lambda x: '%.2f' % x) # truco para notación cientifica
import requests as rq
import datetime
```

### Traemos la tabla "RAWDATA"

```python
df_rawdata = pd.read_csv('raw_data.csv')
df_rawdata
```

### Traemos la tabla "Competidores"

```python
df_competidores = pd.read_csv('Super_competitors .csv')
df_competidores
```

### Usamos la función Merge para unir las tablas "competidores" y "rawdata"

```python
## Join de tablas competidores y parte de rawdata
df_data = pd.merge(df_rawdata, df_competidores,how = 'left', left_on = 'id_reference_supermarket', right_on = 'ID')
display(df_data)
```

### Creamos la tabla "productos"

Con el fin de reducir el dataset a la cantidad de columnas necesarias para nuestro análisis.

```python
df_productos = df_data[['Name','category','name','price','insert_date','product_id']]
df_productos
```

```python
df_productos.dtypes
```

#### Debemos cambiar el tipo de dato de la columna "insert_date"

La columna que contiene las fechas no es del tipo adecuado, por lo que debemos reconvertirla en un formato adecuado.

```python
df_productos['insert_date'] = pd.to_datetime(df_productos['insert_date'])
df_productos.dtypes
## Esta función no toma las horas, solo fecha
```

#### *Al separar la información por supermercado, podemos filtrar luego los productos pertenecientes a las marcas blancas de los mismos.*

### Supermercados Día

Buscamos filtar aquellos productos pertenecientes al Supermercado Día:

```python
solodia = df_productos['Name'] == 'Dia'
dia = df_productos[solodia]
dia
```

Filtramos aquellos productos que se encuentran bajo la marca blanca de Día

```python
blancadia=df_productos[df_productos['name'].str.contains('DIA', na=False)]
blancadia
```

 *De los 3200969 productos (obtenidos en el dataframe dia), hay un total de 783543 registros de su marca blanca.*
 *En este caso, estamos viendo productos repetidos. Ya que estarían los id de cada producto que se generan todos los días de cada año, repitiéndose el mismo producto. Debemos hallar valores únicos.*

#### Valores únicos para Dia

```python
uniqdia = pd.DataFrame(blancadia['product_id'].unique())
uniqdia
```
